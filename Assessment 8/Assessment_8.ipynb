#### 1. Student Grade Calculator
     Write a function calculate_grade(marks) that takes marks as input and returns grade according to:

      90 and above → A
      80 -- 89 → B
      70 -- 79 → C
      Below 70 → Fail

```python
def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif 80<=marks<89:
        return "B"
    elif 70<=marks<79:
        return "C"
    else:
        return "Fail"
marks = int(input("Enter your marks: "))
grade = calculate_grade(marks)
print(grade)
```

### 2. Even or Odd Number Checker
         Write a function is_even(number) that returns:
                 True if the number is even
                 False if the number is odd

```python
def is_even(number):
    if number % 2 ==0 :
        return True
    else:
        return False
number = int(input("Enter number: "))
ans = is_even(number)
print(ans)
```

### 3. Factorial Using Function
         Write a function factorial(n) that returns the factorial of a number using a loop.
         Example: factorial(5) 
                  Output → 120

```python
def factorial(n):
    result = 1 
    for i in range(1, n + 1):
        result *= i  
    return result

num = int(input("Enter a number to find factorial: "))
print(f"Factorial of {num} is {factorial(num)}")
```

### 4. Find Maximum Number
       Write a function find_max(a, b, c) that returns the largest number among three numbers.

```python
def find_max(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

# Example usage
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

print(f"The maximum number is: {find_max(num1, num2, num3)}")
```

### 5. Word Counter
        Write a function count_words(sentence) that counts how many words exist in a given sentence.
       Example: "I love Python programming"
        Output → 4

```python
def count_words(sentence):
    words = sentence.split()  
    return len(words)      

sentence = input("Enter a sentence: ")
print(f"Number of words: {count_words(sentence)}")
```

### 6. Prime Number Checker
     Write a function is_prime(n) that checks whether a number is a prime number.

```python
def is_prime(n):
    if n <= 1:
        return False  
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False  
    return True  
num = int(input("Enter a number: "))
if is_prime(num):
    print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

### 7. List Sum Function
       Write a function list_sum(numbers) that takes a list as input and returns the sum of all numbers.
     Example: [10, 20, 30]
      Output → 60

```python
def list_sum(numbers):
    total = 0
    for num in numbers:
        total += num  
    return total

nums = [10, 20, 30]
print(f"Sum of the list is: {list_sum(nums)}")
```

# 8. Password Strength Checker
     Write a function check_password(password) that checks:
        Minimum 8 characters required
        Must contain at least one digit
         Return: "Strong" or "Weak"

```python
def check_password(password):
    # Check minimum length
    if len(password) < 8:
        return "Weak"
    
    # Check if it contains at least one digit
    has_digit = False
    for char in password:
        if char.isdigit():
            has_digit = True
            break
    
    if not has_digit:
        return "Weak"
    
    # Passed all checks
    return "Strong"

# Example usage
pwd = input("Enter your password: ")
print(f"Password is {check_password(pwd)}")
```

### 9. Temperature Converter
      Write a function celsius_to_fahrenheit(c) that converts temperature from Celsius to Fahrenheit.
        Formula: F = (C × 9/5) + 32

```python
def celsius_to_fahrenheit(c):
    f = (c * 9/5) + 32
    return f

celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = celsius_to_fahrenheit(celsius)
print(f"{celsius}°C is equal to {fahrenheit}°F")
```

### 10. Remove Duplicates from List
       Write a function remove_duplicates(data) that removes duplicate elements from a list and returns a new list.
       Example: [1,2,2,3,4,4]
        Output → [1,2,3,4]

```python
def remove_duplicates(data):
    new_list = []
    for item in data:
        if item not in new_list:
            new_list.append(item)
    return new_list
    
data = [1, 2, 2, 3, 4, 4]
print(f"List after removing duplicates: {remove_duplicates(data)}")
```
