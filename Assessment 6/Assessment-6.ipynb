## 3. Create a variable “Grade” holding an integer between 0 to100. Code to print
the letter grade of the number grade variable.

Grades:

- A = 90 - 100
 
- B = 80 - 89

- C = 70-79

- D = 60 - 69
 
- F = 0 – 59

```python
grade = int(input("Enter integer between 0 to 100 : "))
if 90 <= grade <= 100:
    print('A')
elif 80 <= grade <= 89:
    print('B')
elif 70 <= grade <= 79:
    print('C')
elif 60 <= grade <= 69:
    print('D')
else:
    print('F')
```
